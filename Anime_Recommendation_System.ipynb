```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
movie['tags'] = movie['tags'].apply(lambda x:x.lower())
```

```python
movie.rename(columns = {'name':'title'}, inplace = True)
```

```python
movie
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
movie['tags'].apply(stem)
```

```python
movie.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words ='english')
```

```python
vectors = cv.fit_transform(movie['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(mov):
    index = movie[movie['title'] == mov].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    for i in distances[1:6]:
        print(movie.iloc[i[0]].title)
```

```python
recommend('Death Note')
```
